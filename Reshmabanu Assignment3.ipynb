```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='b',label='setosa')
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='b',label='setosa')
sns.kdeplot(iris.loc[(iris['species']=='versicolor'),'sepal_length'],color='r',label='versi')
```

```python

sns.relplot(data=iris,x="sepal_length",y="sepal_width")
```

```python
sns.relplot(data=iris,x="petal_length",y="petal_width")
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='b',label='setosa')
sns.kdeplot(iris.loc[(iris['species']=='versicolor'),'sepal_length'],color='r',label='versi')
sns.kdeplot(iris.loc[(iris['species']=='virginica'),'sepal_length'],color='g',label='virgin')
```

```python
sns.boxplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
iris=np.random.randn(1000)
plt.hist(iris)
```

```python
labels='car','scooty','truck','bus'
sizes=[225,130,245,210]
colors=['r','b','g','c']
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
